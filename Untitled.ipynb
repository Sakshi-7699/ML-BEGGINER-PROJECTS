```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import matplotlib.pyplot as plt
import numpy as np
```

```python
df= pd.read_csv('winequality-red.csv')
df.head()
```

```python
col=df.columns
print(col)
```

```python
x=df[['fixed acidity', 'volatile acidity', 'citric acid', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']]
y = df['quality']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=3)
```

```python
x_train
```

```python
regressor= LinearRegression()
regressor.fit(x_train,y_train)
```

```python
pre_x=regressor.predict(x_test)
```

```python
pre_x[0]
```

```python
type(y_test)
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, pre_x))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, pre_x))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, pre_x)))
```
